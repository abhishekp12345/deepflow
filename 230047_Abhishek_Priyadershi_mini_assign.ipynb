```python
# convert to string in case it's a mix of types
zomato_dataset['approx_cost(for two people)']=zomato_dataset['approx_cost(for two people)'].astype(str)
# Remove commas and  symbols (if present), then convert to float
zomato_dataset['approx_cost(for two people)'] = zomato_dataset['approx_cost(for two people)'].str.replace(',', '').str.replace('', '').astype(float)
zomato_dataset['approx_cost(for two people)']
```

Q6.Standardise key columns like ratings, price, and location.

```python
# converting the values into numeric form
zomato_dataset['Rating'] = pd.to_numeric(zomato_dataset['Rating'], errors='coerce')
print(zomato_dataset['Rating'])
```

7. Convert columns to appropriate data types (e.g., float, int, datetime).


```python
zomato_dataset['votes']=pd.to_numeric(zomato_dataset['votes'], errors='coerce')
```

8. Save the cleaned and structured dataset for analysis.

```python
zomato_dataset.to_csv('zomato_cleaned.csv', index=False)
```

9 Group data by cuisine and calculate average ratings, votes, and price range.

```python
grouped_data=zomato_dataset.groupby('cuisines').agg({
    'Rating': 'mean',
    'votes': 'mean',
    'approx_cost(for two people)': 'mean'

}).reset_index()
```

10. Identify top cuisines based on ratings and popularity (votes).


```python
top_cuisines=grouped_data.sort_values(by=['Rating','votes'],ascending=False)
top_cuisines.head()
```

11. Highlight best-rated restaurants and link them to their cuisine types.


```python
top_restaurants= zomato_dataset.sort_values(by=['Rating'],ascending=False).groupby('cuisines').head(1)
top_restaurants
```

12. Analyse and generate insights into user preferences by cuisine.


```python

zomato_dataset['cuisines'].value_counts()
```

13. Visualise key findings using Matplotlib.


```python
plt.figure(figsize=(6,3))
plt.scatter(zomato_dataset['approx_cost(for two people)'], zomato_dataset['Rating'], alpha=0.5, color='green')
plt.title('Cost for Two vs Rating')
plt.xlabel('Average Cost for Two')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.scatter(zomato_dataset['votes'], zomato_dataset['Rating'], alpha=0.5, color='coral')
plt.title('Votes vs Rating')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```
